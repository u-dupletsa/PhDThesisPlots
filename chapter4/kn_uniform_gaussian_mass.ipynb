```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
```

```python
# General settings
import matplotlib as mpl
mpl.rc_file('/Users/ulyanadupletsa/Documents/PhDThesisPlots/matplotlibrc')

mpl.rc('xtick', labelsize = 25) 
mpl.rc('ytick', labelsize = 25)
```

```python
my_color = '#97a7a7'
my_color_darker = '#2f4f4f'
my_color2 = '#eba75f'
my_color2_darker = '#f6b26b'
myfontsize=25
```

```python
fig, ax = plt.subplots(1)
plt.hist(np.random.normal(1.33, 0.09, 10_000), bins=50, histtype='step', linestyle='solid', linewidth=2., color=my_color2_darker, label='Gaussian')
plt.hist(np.random.uniform(1.1, 2.2, 10_000), bins=50, histtype='step', linestyle='solid', linewidth=2., color=my_color_darker, label='Uniform')
fig.legend(ncol=2, frameon=False, fontsize=myfontsize, loc='upper center', bbox_to_anchor=(0.5,1.05))
plt.yscale('log')
plt.xlim(0.9, 2.25)
plt.ylabel('Counts', fontsize=myfontsize)
plt.xlabel('$m$ [$M_{\odot}$]', fontsize=myfontsize)
plt.savefig('plots/gaussian_vs_uniform.pdf', bbox_inches='tight')
plt.show()
```
