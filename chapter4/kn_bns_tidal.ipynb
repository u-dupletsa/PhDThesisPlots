```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches 
import matplotlib.colors as mcolors
import seaborn as sns
```

```python
# General settings
import matplotlib as mpl
mpl.rc_file('/Users/ulyanadupletsa/Documents/PhDThesisPlots/matplotlibrc')

mpl.rc('xtick', labelsize = 23) 
mpl.rc('ytick', labelsize = 23)
```

```python
my_color_darker = '#2f4f4f'
my_color2_darker = '#f6b26b'
```

```python
PATH_TO_DATA = '/Users/ulyanadupletsa/Documents/KN/results_10years'
```

```python
lbs_signals = ['mass_1', 'mass_2', 'luminosity_distance', 'dec', 'ra', 'theta_jn',
	   'psi', 'geocent_time', 'phase', 'redshift', 'lambda_1', 'lambda_2', 'network_SNR']
```

```python
alpha=1.0
detector_ntw='ET_delta_10_cryo'
myfontsize=23
```

```python
injections_BLh_gaussian = pd.read_csv('%s/%s/BLh_gaussian/%s/Signals_%s_BLh_gaussian_%s.txt' %(str(PATH_TO_DATA), str(alpha), 
					str(detector_ntw), str(alpha), str(detector_ntw)),
						names = lbs_signals, skiprows = 1, delimiter = ' ')
injections_APR4_gaussian = pd.read_csv('%s/%s/APR4_gaussian/%s/Signals_%s_APR4_gaussian_%s.txt' %(str(PATH_TO_DATA), str(alpha), 
					str(detector_ntw), str(alpha), str(detector_ntw)),
						names = lbs_signals, skiprows = 1, delimiter = ' ')

injections_BLh_uniform = pd.read_csv('%s/%s/BLh_uniform/%s/Signals_%s_BLh_uniform_%s.txt' %(str(PATH_TO_DATA), str(alpha), 
					str(detector_ntw), str(alpha), str(detector_ntw)),
						names = lbs_signals, skiprows = 1, delimiter = ' ')
injections_APR4_uniform = pd.read_csv('%s/%s/APR4_uniform/%s/Signals_%s_APR4_uniform_%s.txt' %(str(PATH_TO_DATA), str(alpha), 
					str(detector_ntw), str(alpha), str(detector_ntw)),
						names = lbs_signals, skiprows = 1, delimiter = ' ')
fig, ax = plt.subplots(1)
plt.hist(injections_BLh_uniform['lambda_1'], bins=30, histtype='step', linestyle='solid', linewidth=2.5, color=my_color_darker, label='BLh-Uniform')
plt.hist(injections_APR4_uniform['lambda_1'], bins=30, histtype='step', linestyle='dashed', linewidth=2.5, color=my_color_darker, label='APR4-Uniform')
plt.hist(injections_BLh_gaussian['lambda_1'], bins=30, histtype='step', linestyle='solid', linewidth=2.5, color=my_color2_darker, label='BLh-Gaussian')
plt.hist(injections_APR4_gaussian['lambda_1'], bins=30, histtype='step', linestyle='dashed', linewidth=2.5, color=my_color2_darker, label='APR4-Gaussian')

fig.legend(ncol=1, frameon=False, fontsize=myfontsize, loc='upper center', bbox_to_anchor=(0.5,1.35))
plt.yscale('log')
plt.ylabel('Counts', fontsize=myfontsize+3)
plt.xlabel('$\Lambda_1$', fontsize=myfontsize+3)
plt.savefig('plots/tidal_def_kn.pdf', bbox_inches='tight')
plt.show()
```
