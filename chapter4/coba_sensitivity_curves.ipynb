```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# General settings
import matplotlib as mpl
mpl.rc_file('/Users/ulyanadupletsa/Documents/PhDThesisPlots/matplotlibrc')

mpl.rc('xtick', labelsize = 16) 
mpl.rc('ytick', labelsize = 16)
```

```python
my_color = '#97a7a7'
my_color_darker = '#2f4f4f'
my_color2 = '#eba75f'
my_color2_darker = '#f6b26b'
```

```python
PATH_TO_PSDS = '/Users/ulyanadupletsa/Documents/CoBA/PSDs/'
```

```python
et_D = pd.read_csv(PATH_TO_PSDS + 'ET_D_psd.txt', delimiter = ' ', names = ['f', 'PSD'])

et_cryo10 = pd.read_csv(PATH_TO_PSDS + 'ET_10_full_cryo_psd.txt', delimiter = ' ', names = ['f', 'PSD'])
et_cryo15 = pd.read_csv(PATH_TO_PSDS + 'ET_15_full_cryo_psd.txt', delimiter = ' ', names = ['f', 'PSD'])
et_cryo20 = pd.read_csv(PATH_TO_PSDS + 'ET_20_full_cryo_psd.txt', delimiter = ' ', names = ['f', 'PSD'])

et_hf10 = pd.read_csv(PATH_TO_PSDS + 'ET_10_HF_psd.txt', delimiter = ' ', names = ['f', 'PSD'])
et_hf15 = pd.read_csv(PATH_TO_PSDS + 'ET_15_HF_psd.txt', delimiter = ' ', names = ['f', 'PSD'])
et_hf20 = pd.read_csv(PATH_TO_PSDS + 'ET_20_HF_psd.txt', delimiter = ' ', names = ['f', 'PSD'])
```

```python
## One only plot

fig, axs = plt.subplots(1, 2, figsize=(10, 4), sharex=False, sharey=True)

# Plot cryo asds on a loglog plot
ax = axs[0]
ax.loglog(et_D['f'], np.sqrt(et_D['PSD']), label='ET-D', color='black', linewidth=2)

ax.loglog(et_cryo10['f'], np.sqrt(et_cryo10['PSD']), label='ET 10km cryo', color=my_color_darker, alpha=0.4, linewidth=2)
ax.loglog(et_cryo15['f'], np.sqrt(et_cryo15['PSD']), label='ET 15km cryo', color=my_color_darker, alpha=0.65, linewidth=2)
ax.loglog(et_cryo20['f'], np.sqrt(et_cryo20['PSD']), label='ET 20km cryo', color=my_color_darker, alpha=0.9, linewidth=2)

ax.set_xlabel('f [Hz]', fontsize=18)
ax.set_ylabel('ASD [Hz$^{-1/2}$]', fontsize=18)
ax.legend(fontsize=16, frameon=False)
ax.set_title('HFLF-cryo', fontsize=21, weight='bold')

# Plot HF asds on a loglog plot
ax = axs[1]
ax.loglog(et_D['f'], np.sqrt(et_D['PSD']), label='ET-D', color='black', linewidth=2)

ax.loglog(et_hf10['f'], np.sqrt(et_hf10['PSD']), label='ET 10km HF', color=my_color2_darker, alpha=0.4, linewidth=2)
ax.loglog(et_hf15['f'], np.sqrt(et_hf15['PSD']), label='ET 15km HF', color=my_color2_darker, alpha=0.65, linewidth=2)
ax.loglog(et_hf20['f'], np.sqrt(et_hf20['PSD']), label='ET 20km HF', color=my_color2_darker, alpha=0.9, linewidth=2)

ax.set_xlabel('f [Hz]', fontsize=18)
ax.legend(fontsize=16, frameon=False)
ax.set_title('HF', fontsize=21, weight='bold')

plt.tight_layout()
plt.savefig('plots/coba_sens_curves.pdf', bbox_inches='tight')
plt.show()
```
