```python
import numpy as np
import astropy
import matplotlib
import matplotlib.pyplot as plt
```

```python
# General settings
import matplotlib as mpl
mpl.rc_file('/Users/ulyanadupletsa/Documents/PhDThesisPlots/matplotlibrc')

mpl.rc('xtick', labelsize = 23) 
mpl.rc('ytick', labelsize = 23)
```

```python
my_color = '#97a7a7'
my_color_darker = '#2f4f4f'
my_color2 = '#f8c997'
my_color2_darker = '#f6b26b'
```

```python
from astropy.cosmology import FlatLambdaCDM

# Cosmological parameters (H0 in km/s/Mpc, Omega_m, Omega_Lambda)
H0 = 67.77  # Hubble constant (km/s/Mpc)
Omega_m = 0.31  # Matter density parameter
h=H0/100

# Create a FlatLambdaCDM cosmology model
cosmo = FlatLambdaCDM(H0=H0, Om0=Omega_m)

# Redshift range
z_values = np.linspace(0, 3, 500)

# Compute the comoving distance for each redshift value
D_values = cosmo.comoving_distance(z_values)

# Plotting the comoving distance vs redshift
plt.figure(figsize=(8, 6))
plt.loglog(z_values, D_values*h, color=my_color_darker, linewidth=2.5)
plt.axhline(500, color=my_color2_darker, linestyle='dashed')
plt.text(0.005, 520, 'Box Size', fontsize=23, color=my_color2_darker)
plt.xlabel('$z$', fontsize=26)
plt.ylabel('Comoving Distance [Mpc/$h$]', fontsize=26)
plt.grid(True)
plt.savefig('plots/com_distance.pdf', bbox_inches='tight')
plt.show()
```
