```python
my_color = '#97a7a7'
my_color_darker = '#2f4f4f'
my_color2 = '#eba75f'
my_color2_darker = '#f6b26b'
my_fontsize = 25
```

```python
psd_data_LIGO = np.loadtxt('data/psd_IMRPhenomXPHM_IGWN-GWTC2p1-v2-GW150914_095045_PEDataRelease_mixed_cosmo_H1.txt')
psd_data_ET = gw.utilities.get_detector_psd('ET')
```

```python
plt.loglog(psd_data_LIGO[:, 0], np.sqrt(psd_data_LIGO[:, 0] * psd_data_LIGO[:, 1]), color=my_color2_darker, linewidth=2.5, label='LIGO Hanford')
plt.loglog(psd_data_ET[:, 0], np.sqrt(psd_data_ET[:, 0] * psd_data_ET[:, 1]), color=my_color_darker, linewidth=2.5, label='ET_D')
plt.legend(fontsize=23, frameon=False, loc='upper right')
plt.ylim(1e-24, 1e-16)
plt.xlim(1, 6000)
plt.xlabel('Frequency [Hz]', fontsize=my_fontsize)
plt.ylabel('Characteristic Strain', fontsize=my_fontsize)
plt.savefig('plots/sensitivity_cfr.pdf', bbox_inches='tight')
plt.show()
```
