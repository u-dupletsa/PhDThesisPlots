```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
# General settings
import matplotlib as mpl
mpl.rc_file('/Users/ulyanadupletsa/Documents/PhDThesisPlots/matplotlibrc')

mpl.rc('xtick', labelsize = 21) 
mpl.rc('ytick', labelsize = 21)
```

```python
my_color = '#2f4f4f'
my_color2 = '#f8c997'
my_color2_darker = '#f6b26b'
```

```python
data_from_paper = pd.read_csv('data/Figure1.csv', delimiter='\t')
```

```python
# plot the H0 posterior
fig, ax = plt.subplots(figsize=(6, 4))
ax.axvspan(67.27-0.60, 67.27+0.60, color=my_color, alpha=0.5, label='Planck')
ax.axvspan(73.2-1.3, 73.2+1.3, color=my_color2, alpha=0.5, label='SH$_0$ES')
ax.axvline(67.27, color=my_color, linestyle='solid', lw=1.5)
ax.axvline(73.2, color=my_color2, linestyle='solid', lw=1.5)
sns.kdeplot(data_from_paper['H0_samples'].to_numpy(), color='black', linewidth=1.5, label='GW170817')  
plt.xlabel('$H_0$ [km s$^{-1}$ Mpc$^{-1}$]', fontsize=23)
plt.ylabel('P($H_0 |$GW170817)', fontsize=23)
plt.legend(fontsize=21, frameon=False)
plt.savefig('plots/H0_posterior_gw170817.pdf', bbox_inches='tight')
plt.show()
```
