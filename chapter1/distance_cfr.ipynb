```python
my_color = '#97a7a7'
my_color_darker = '#2f4f4f'
my_color2 = '#f8c997'
my_color2_darker = '#f6b26b'
```

```python
from astropy.cosmology import Planck18

# Redshift range
zz = np.linspace(0.01, 100, 1_000)

# Compute the comoving distance for each redshift value
comoving_distance = Planck18.comoving_distance(zz)

# Compute the angular diameter distance for each redshift value
angular_diameter_distance = Planck18.angular_diameter_distance(zz)

# Compute the luminosity distance for each redshift value
luminosity_distance = Planck18.luminosity_distance(zz)

# Plotting the comoving distance vs redshift
plt.figure(figsize=(8, 6))
plt.loglog(zz, luminosity_distance, color=my_color_darker, linestyle='dashed', linewidth=2.5, label='$d_L(z)$')
plt.loglog(zz, comoving_distance, color=my_color_darker, linewidth=2.5, label='$d_{\\rm comoving}$')
plt.loglog(zz, angular_diameter_distance, color=my_color_darker, linestyle='dotted', linewidth=2.5, label='$d_A(z)$')
plt.xlabel('$z$', fontsize=26)
plt.ylabel('Distance [Mpc]', fontsize=26)
plt.legend(frameon=False, fontsize=26)
plt.savefig('plots/cfr_distance.pdf', bbox_inches='tight')
plt.show()
```
