```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

```python
# General settings
import matplotlib as mpl
mpl.rc_file('/Users/ulyana/Documents/PhDThesisPlots/matplotlibrc')

mpl.rc('xtick', labelsize = 21) 
mpl.rc('ytick', labelsize = 21)
```

```python
path_to_plots = 'plots/'
```

```python
my_color = '#2f4f4f'
```

```python
path_to_inj = '/home/ulyana.dupletsa/MDC-Paper1/data/injections_vanilla_case.hdf5'
vanilla_inj = pd.read_hdf(path_to_inj)
select_params = ['m1s', 'm2s', 'dls']
vanilla_inj_sel = vanilla_inj[select_params]

vanilla_inj_df = pd.DataFrame.from_dict({
                    'mass_1_source':vanilla_inj_sel['m1s'].loc[:],
                    'mass_2_source':vanilla_inj_sel['m2s'].loc[:],
                    'luminosity_distance':vanilla_inj_sel['dls'].loc[:]
                    })
```

```python
my_bin_num = 30
```

### Vanilla injections

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharex = False, sharey=True)

sns.histplot(vanilla_inj_df['mass_1_source'], bins = my_bin_num, fill = True, stat = 'count',
                    color = my_color, alpha=0.5,
                    linewidth = 0.75, ax = axs[0],
                    edgecolor=my_color)
axs[0].set_xlim(0, 100)
axs[0].set_xlabel('$m_{\\rm 1,source}$ [$M_{\odot}$]', fontsize=25)
axs[0].set_ylabel('Counts', fontsize=25)

sns.histplot(vanilla_inj_df['mass_2_source'], bins = my_bin_num, fill = True, stat = 'count',
                    color = my_color, alpha=0.5,
                    linewidth = 0.75, ax = axs[1],
                    edgecolor=my_color)
axs[1].set_xlim(0, 100)
axs[1].set_xlabel('$m_{\\rm 2,source}$ [$M_{\odot}$]', fontsize=25)

sns.histplot(vanilla_inj_df['luminosity_distance'], bins = my_bin_num, 
                    fill = True, stat = 'count',
                    color = my_color, alpha=0.5,
                    linewidth = 0.75, ax = axs[2],
                    edgecolor=my_color)
axs[2].set_xlabel('$d_L$ [Mpc]', fontsize=25)


plt.subplots_adjust(wspace=0.1, 
                    hspace=0.2)

plt.savefig(path_to_plots + '/' + 'vanilla_injections.pdf', bbox_inches='tight')
plt.show()
```
